## 定義 Gemini 餐廳搜尋功能

### Subtask:
建立一個 Python 函式，該函式接收使用者的自然語言查詢（例如：「附近的義式餐廳」），然後使用 Gemini API 解析查詢，提取出關鍵資訊，例如餐廳類型和大致的地理位置意圖。


### 定義 Gemini 餐廳搜尋功能

此函式將接收使用者的自然語言查詢，並利用 Gemini API 解析出關鍵的餐廳類型和地點意圖，以 JSON 格式返回。


**Reasoning**:
I will implement the `gemini_restaurant_search` function, which uses the Gemini API to parse natural language queries, extract restaurant type and location intent, and return them in a dictionary format as specified in the subtask.



````python
import google.generativeai as genai
import json
import os

def gemini_restaurant_search(query: str) -> dict:
    """
    Parses a natural language query using the Gemini API to extract restaurant type and location intent.

    Args:
        query: The user's natural language query (e.g., "附近的義式餐廳").

    Returns:
        A dictionary containing 'restaurant_type' and 'location_intent',
        e.g., {'restaurant_type': '義式', 'location_intent': '我目前位置附近'}.
        Returns an empty dictionary or error details if parsing fails.
    """
    GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
    if not GEMINI_API_KEY or GEMINI_API_KEY == "YOUR_GEMINI_API_KEY":
        # Use a print statement instead of returning directly to allow error propagation to Gradio output
        print("Error: GEMINI_API_KEY is not set or is a placeholder. Please set your actual Google Gemini API key.")
        return {"error": "GEMINI_API_KEY not set"}

    try:
        genai.configure(api_key=GEMINI_API_KEY)

        # Dynamically select a model that supports generateContent
        model_name = None
        for m in genai.list_models():
            if 'generateContent' in m.supported_generation_methods:
                if m.name == 'models/gemini-pro': # Prioritize gemini-pro if available and supported
                    model_name = m.name
                    break
                elif model_name is None: # Otherwise, pick the first supported model
                    model_name = m.name

        if not model_name:
            return {"error": "No Gemini model found that supports generateContent."}

        model = genai.GenerativeModel(model_name)

        prompt = f"""
        請從以下使用者查詢中提取「餐廳類型」(restaurant_type) 和「地點意圖」(location_intent)。
        以 JSON 格式返回，例如: {{"restaurant_type": "...", "location_intent": "..."}}。
        如果無法確定，請使用空字串。地點意圖可以是具體的地址、地標或模糊的描述（如「我目前位置附近」、「市中心」）。

        使用者查詢: "{query}"
        """

        response = model.generate_content(prompt)

        # Extract the text content from the response
        json_string = response.text.strip()

        # Attempt to parse the JSON string
        parsed_data = json.loads(json_string)

        restaurant_type = parsed_data.get("restaurant_type", "")
        location_intent = parsed_data.get("location_intent", "")

        return {"restaurant_type": restaurant_type, "location_intent": location_intent}

    except Exception as e:
        # Ensure error message is captured for Gradio output
        return {"error": str(e)}

print("Defined gemini_restaurant_search function.")
````

## 定義 Google Maps 餐廳詳細資訊功能

### Subtask:
建立一個 Python 函式，該函式接收從 Gemini 提取出的餐廳類型和位置資訊，然後使用 Google Maps Places API 執行搜尋。此函式應負責從 Places API 結果中提取餐廳名稱、地址、評分、評論摘要以及可能的 Google Maps 連結。


**Reasoning**:
I need to implement the `google_maps_restaurant_details` function as described in the subtask. This involves using the `googlemaps` client to perform a place search, extract specific details from the results, and format them into a list of dictionaries.



````python
import googlemaps
import os

def google_maps_restaurant_details(restaurant_type: str, location_intent: str) -> list:
    """
    Searches for restaurant details using Google Maps Places API based on restaurant type and location intent.

    Args:
        restaurant_type: The type of restaurant to search for (e.g., "義式", "日式").
        location_intent: The intended location for the search (e.g., "我目前位置附近", "市中心", "台北車站").

    Returns:
        A list of dictionaries, where each dictionary contains details for a restaurant
        (name, address, rating, reviews_total, map_link).
        Returns an empty list if no results are found.
        Returns a dictionary with an 'error' key if an error occurs or API key is missing.
    """
    GOOGLE_MAPS_API_KEY = os.getenv("GOOGLE_MAPS_API_KEY")
    if not GOOGLE_MAPS_API_KEY or GOOGLE_MAPS_API_KEY == "YOUR_GOOGLE_MAPS_API_KEY":
        print("Error: GOOGLE_MAPS_API_KEY is not set or is a placeholder. Please set your actual Google Maps API key.")
        return {"error": "GOOGLE_MAPS_API_KEY not set"}

    try:
        # Re-initialize gmaps client inside the function for robustness, or assume it's global
        # For this context, we will re-initialize to ensure it's always ready.
        gmaps = googlemaps.Client(key=GOOGLE_MAPS_API_KEY)

        # Construct the query string
        query = f"{location_intent} {restaurant_type} 餐廳"

        # Perform a places search. Using 'text search' for broader queries.
        places_result = gmaps.places(query=query, language='zh-TW')

        restaurants = []
        for place in places_result.get('results', []):
            name = place.get('name', 'N/A')
            address = place.get('formatted_address', 'N/A')
            rating = place.get('rating', 0.0)
            user_ratings_total = place.get('user_ratings_total', 0)
            place_id = place.get('place_id', '')

            # Construct Google Maps link
            map_link = f"https://www.google.com/maps/search/?api=1&query={name},{address}&query_place_id={place_id}"

            restaurants.append({
                'name': name,
                'address': address,
                'rating': rating,
                'reviews_total': user_ratings_total,
                'map_link': map_link
            })
        return restaurants

    except Exception as e:
        print(f"An error occurred during Google Maps API call: {e}")
        return {"error": str(e)}

print("Defined google_maps_restaurant_details function.")
````

**Reasoning**:
Now that the basic Gradio interface is defined, I will create a new function that integrates the `gemini_restaurant_search` and `google_maps_restaurant_details` functions. This function will serve as the core logic for the Gradio application, taking a user query, parsing it with Gemini, searching with Google Maps, and formatting the results for display.



````python
import gradio as gr
import os
import traceback # Import traceback module

last_gradio_output = ""

def restaurant_finder_app(query: str) -> str:
    global last_gradio_output # Declare intent to modify global variable
    debug_info = f"### 接收到查詢: {query}\n\n"
    try:
        # 1. Parse query using Gemini API
        gemini_result = gemini_restaurant_search(query)
        debug_info += f"**Gemini API 解析結果:** `{gemini_result}`\n\n"

        if "error" in gemini_result:
            last_gradio_output = debug_info + f"錯誤：Gemini API 呼叫失敗 - {gemini_result['error']}"
            return last_gradio_output

        restaurant_type = gemini_result.get('restaurant_type')
        location_intent = gemini_result.get('location_intent')

        if not restaurant_type and not location_intent:
            last_gradio_output = debug_info + "我無法從您的查詢中理解餐廳類型或地點意圖。請嘗試更明確地描述，例如：'台北市信義區的義式餐廳'。"
            return last_gradio_output

        debug_info += f"**提取出的意圖:** 餐廳類型: `{restaurant_type}`, 地點意圖: `{location_intent}`\n\n"

        # 2. Search for restaurants using Google Maps API
        maps_results = google_maps_restaurant_details(restaurant_type, location_intent)
        debug_info += f"**Google Maps API 搜尋結果:** `{maps_results}`\n\n"

        if isinstance(maps_results, dict) and "error" in maps_results:
            last_gradio_output = debug_info + f"錯誤：Google Maps API 呼叫失敗 - {maps_results['error']}"
            return last_gradio_output

        if not maps_results:
            last_gradio_output = debug_info + f"抱歉，在 '{location_intent}' 附近沒有找到 '{restaurant_type}' 餐廳。"
            return last_gradio_output

        # 3. Format results for Gradio display
        formatted_results = []
        for restaurant in maps_results:
            name = restaurant.get('name', 'N/A')
            address = restaurant.get('formatted_address', 'N/A')
            rating = restaurant.get('rating', 'N/A')
            reviews_total = restaurant.get('user_ratings_total', 'N/A')
            place_id = restaurant.get('place_id', '')
            map_link = f"https://www.google.com/maps/search/?api=1&query={name},{address}&query_place_id={place_id}"

            formatted_results.append(
                f"#### {name}\n"
                f"地址: {address}\n"
                f"評分: {rating} ({reviews_total} 則評論)\n"
                f"[Google Maps]({map_link})\n"
                "---\n"
            )

        final_output = "".join(formatted_results)
        last_gradio_output = debug_info + "### 搜尋結果:\n\n" + final_output
        return last_gradio_output

    except Exception as e:
        error_traceback = traceback.format_exc()
        last_gradio_output = debug_info + f"### 發生未預期的錯誤！\n\n錯誤類型: {type(e).__name__}\n錯誤訊息: {e}\n\n**追蹤資訊 (Traceback):**\n```\n{error_traceback}\n```"
        return last_gradio_output


# Re-create the Gradio interface with the new integrated function
query_input = gr.Textbox(
    label="輸入您的餐廳查詢",
    placeholder="例如：台北市信義區的日式餐廳",
    lines=2
)

output_display = gr.Markdown(
    label="餐廳搜尋結果"
)

iface = gr.Interface(
    fn=restaurant_finder_app, # Use the new integrated function
    inputs=query_input,
    outputs=output_display,
    title="餐廳搜尋助手",
    description="使用自然語言查詢來尋找餐廳。"
)

# Launch the interface
# Uncomment the line below to run the Gradio interface.
# iface.launch(share=True)

print("Integrated restaurant_finder_app and updated Gradio interface definition with debugging prints.")
````
